```python
# Importing Packages 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parameters for pi, pa, and the coordinates for point p(b)
pi = .5
pa = .4
pb_x, pb_y = 0.7, 0.25  # Coordinates for p(b)
gamma = .5

# Define the range for x values (10000 points between 0 and 1)
x = np.linspace(0, 1, 10000)

# Define the two functions y1 and y2 for plotting, and y3 for the 45-degree line
y1 = (pa) / pi - x  # Grey line function
y2 = (.25 / pi * pa**2 - 0.25 * x**2)**(1 - gamma) / (1 - gamma)  # Blue curve function
y3 = x  # 45-degree line (y=x)

# Plotting the curves and the 45-degree line
plt.figure(figsize=(8, 6))  # Set figure size
plt.plot(x, y1, color='grey')  # Plot the grey curve (y1)
plt.plot(x, y2, color='#1F77B4')  # Plot the blue curve (y2)
plt.plot(x, y3, color='black', linestyle=':', linewidth=1)  # Plot the 45-degree line (dashed)

# Adding small vertical line segments stemming from the curves (y1 and y2)
segment_length = 0.015  # Length of the small vertical segments
plt.vlines(x[::150], y1[::150], y1[::150] + segment_length, color='grey', linestyle='-', alpha=0.6)  # Grey segments on y1
plt.vlines(x[::150], y2[::150], y2[::150] + segment_length, color='#1F77B4', linestyle='-', alpha=0.6)  # Blue segments on y2

# Set limits to only display Quadrant 1 with aesthetics purpose
plt.xlim(0, 1.1)
plt.ylim(0, 1.1)

# Remove the gridlines for simplicity
plt.grid(False)

# Add y-axis tick at 1
plt.yticks([1], ["1"])

# Manually add the label for p^2 without a tick mark for aesthetics
plt.text(-.025, 1.12, "$p^{2}$", fontsize=12, verticalalignment='center')  

# Add x-axis tick at 1
plt.xticks([1], ['1'])

# Manually add the label for p^1 without a tick mark for aesthetics
plt.text(1.12, 0-.025, "$p^{1}$", fontsize=12, horizontalalignment='center')  # Label for p^1 at x=1

# Label the points p(a) and p(b)
plt.plot(pa, pa, 'ko')  # Plot p(a) as a black dot
plt.text(pa - 0.03, pa - 0.04, '$p(a)$', fontsize=10, verticalalignment='center')  # Label for p(a)

plt.plot(pb_x, pb_y, 'o', color='#FFCC00')  # Plot p(b) as a yellow dot
plt.text(pb_x - 0.02, pb_y + 0.04, '$p(b)$', fontsize=10, verticalalignment='center')  # Label for p(b)

# Add line segment connecting p(a) to p(b) in yellow color
plt.plot([pa, pb_x], [pa, pb_y], '#FFCC00', linewidth=2)  # Line from p(a) to p(b)

# Remove the outer box (top and right spines)
ax = plt.gca()  # Get current axes (to modify the spines)
ax.spines['top'].set_visible(False)  # Hide the top spine
ax.spines['right'].set_visible(False)  # Hide the right spine

plt.savefig('Figure1_1.png', format='png')

# Display the plot
plt.show()
```
